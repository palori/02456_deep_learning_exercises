#### Neural network

In this example, we will be working on a subset of the MNIST dataset. Please note that the selected subset contains a total of 7000 samples vs. 2000 samples in the guided assignment. Distribution is 5000/1000/1000 in order to achieve a sufficiently accurate performance estimate.

Our neural network will have 3 hidden layers with 512 units each. Each layer utilises the ReLU activation function and our parameters are optimised using the highly efficient Adam optimizer. The number of epochs is controlled by a simple implementation of the early stopping algorithm using a patience of 10.

We will be running a model with and without 20% dropout between each layer to test if there is a significant difference in performance.

Network setup

```python
class Net(nn.Module):
    
    def __init__(self, num_features, num_units, num_output, dropout):
        super(Net, self).__init__()
        self.dropout = dropout
        
        def add_layer(num_out, num_in):
            return Parameter(init.kaiming_normal_(torch.Tensor(num_out, num_in))), Parameter(init.constant_(torch.Tensor(num_out), 0))
        
        self.W1, self.b1 = add_layer(num_units, num_features)       
        self.W2, self.b2 = add_layer(num_units, num_units)
        self.W3, self.b3 = add_layer(num_units, num_units)
        self.W4, self.b4 = add_layer(num_output, num_units)
        self.activation = torch.nn.ReLU()
        self.lineardrop = nn.Dropout(p=0.2)
        
    def forward(self, x):
        x = F.linear(x, self.W1, self.b1)
        x = self.activation(x)
        
        if self.dropout:
            x = self.lineardrop(x)
        
        x = F.linear(x, self.W2, self.b2)
        x = self.activation(x)
        
        if self.dropout:
            x = self.lineardrop(x)
        
        x = F.linear(x, self.W3, self.b3)
        x = self.activation(x)
        
        if self.dropout:
            x = self.lineardrop(x)
            
        x = F.linear(x, self.W4, self.b4)
        return F.softmax(x, dim=1)
```

Framework setup

```python
class Model:
    def __init__(self, num_units = 512, dropout = False, num_epochs = 200, batch_size = 100, patience = 10, verbose = True):
        self.net = Net(num_features, num_units, num_classes, dropout)
        self.num_epochs = num_epochs
        self.patience = patience
        self.batch_size = batch_size
        self.optimizer = optim.Adam(self.net.parameters())
        self.loss = nn.CrossEntropyLoss()
        self.verbose = verbose
        self.trn_acc, self.val_acc, self.losses = [],[],[]
        
    def evaluate(self, xval, yval):
        num_batches = xval.shape[0]//self.batch_size
        subset = lambda i, size: range(i * size, (i + 1) * size)
        preds,targs = [],[]
        for i in range(num_batches):
            s = subset(i, self.batch_size)
            out = self.net(Variable(torch.from_numpy(xval[s])))
            y = torch.max(out,1)[1]
            targs += list(yval[s])                
            preds += list(y.data.numpy())
        return preds, targs
        
    def train(self, xtrn, ytrn, xval, yval):
        num_batches_trn = xtrn.shape[0]//self.batch_size
        num_batches_val = xval.shape[0]//self.batch_size
        subset = lambda i, size: range(i * size, (i + 1) * size)
        stalecount = 0
        maxacc = 0
        for epoch in range(self.num_epochs):
            cur_loss = 0
            self.net.train()
            
            for i in range(num_batches_trn):
                s = subset(i, self.batch_size)
                y = self.net(Variable(torch.from_numpy(xtrn[s])))
                yhat = Variable(torch.from_numpy(ytrn[s]).long())
                loss = self.loss(y, yhat)
                self.optimizer.zero_grad()
                loss.backward()
                self.optimizer.step()
                cur_loss += loss
            self.losses.append(cur_loss/self.batch_size)
            
            self.net.eval()
            train_preds,train_targs = self.evaluate(xtrn, ytrn)
            val_preds,val_targs = self.evaluate(xval, yval)
            
            self.trn_acc.append(accuracy_score(train_targs, train_preds))
            self.val_acc.append(accuracy_score(val_targs, val_preds))
            
            def print_progress():
                print('Epoch %3i: Train loss %0.4f , Train acc %0.3f, Valid acc %0.3f' % (
                epoch+1, self.losses[-1], self.trn_acc[-1], self.val_acc[-1]))
            
            if self.verbose and (epoch+1) % 10 == 0:
                print_progress()
                
            if epoch > 0 and self.val_acc[-1] <= maxacc:
                stalecount += 1
                if stalecount > self.patience:
                    if self.verbose:
                        print('Exiting on epoch %i due to early stopping...' % (epoch+1))
                    break
            else:
                stalecount = 0
                maxacc = self.val_acc[-1]
                
        if self.verbose:
            self.plot()
            
    def test(self, xtst, ytst):
        preds,targs = self.evaluate(xval, yval)
        acc = accuracy_score(targs,preds)
        if self.verbose:
            print('Test acc: %0.3f' % accuracy_score(targs,preds))
        return acc
                
    def plot(self):
        plt.figure()
        epoch = np.arange(len(self.trn_acc))
        plt.plot(epoch, self.trn_acc, 'r', epoch, self.val_acc, 'b')
        plt.legend(['Train Accucary','Validation Accuracy'])
        plt.xlabel('Updates'), plt.ylabel('Acc')
```

#### Training and testing of models

Training of a model

```python
model = Model()
model.train(xtrn, ytrn, xval, yval)
```

Testing of a model

```python
acc = model.test(xtst, ytst)
```

Training of a model w/ dropout

```python
model2 = Model(dropout=True)
model2.train(xtrn, ytrn, xval, yval)
```

Testing of a model w/ dropout

```python
acc2 = model2.test(xtst, ytst)
```

As can be seen from the two tests, there is no gain in performance from adding dropout to the dense layers. Any difference in performance can be attributed to randomness. It is however worth noting that dropout adds substantially to training time due to an increased number of epochs required before converging.

It can be postulated that adding dropout on a layer with spacial information (convolutional layers) may have a larger impact on the performance than what could be shown in this example, but I guess we will find out next week ;)

#### Chapter 1 - Bitwise representation of a digit

Recall that 0x000 = 0, 0x001 = 1, 0x011 = 2, etc...

In that case, a bit-conversion layer can be embedded to the output of the output layer. As it can be assumed that predictions are clustered around 0.0 and 1.0, mapping large weights onto edges that may trigger a bit-flip, e.g. the edge from output node corresponding to integer 1 to the node representing the right-most bit. In the opposite case, mapping a large negative weight onto edges will result in negative predicition when passed through an (e.g. sigmoidal) activation function.
